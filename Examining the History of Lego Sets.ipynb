## The Data

You have been provided with two datasets to use. A summary and preview are provided below.

## lego_sets.csv

| Column     | Description              |
|------------|--------------------------|
| `"set_num"` | A code that is unique to each set in the dataset. This column is critical, and a missing value indicates the set is a duplicate or invalid! |
| `"name"` | The name of the set. |
| `"year"` | The date the set was released. |
| `"num_parts"` | The number of parts contained in the set. This column is not central to our analyses, so missing values are acceptable. |
| `"theme_name"` | The name of the sub-theme of the set. |
| `"parent_theme"` | The name of the parent theme the set belongs to. Matches the name column of the parent_themes csv file.
|

## parent_themes.csv

| Column     | Description              |
|------------|--------------------------|
| `"id"` | A code that is unique to every theme. |
| `"name"` | The name of the parent theme. |
| `"is_licensed"` | A Boolean column specifying whether the theme is a licensed theme. |

```python
# Import pandas, read and inspect the datasets
import pandas as pd

lego_sets = pd.read_csv('data/lego_sets.csv')
lego_sets.head()
```

```python
parent_themes = pd.read_csv('data/parent_themes.csv')
parent_themes.head()
```

```python
# Perform an inner join on lego_sets and parent_themes using the 'parent_theme' and 'name' columns
merged_df = pd.merge(lego_sets, parent_themes, how='inner', left_on='parent_theme', right_on='name')

# Display the first few rows of the merged dataframe
merged_df.head()
```

```python
# Filter merged_df to include only rows where is_licensed is True
licensed_df = merged_df[merged_df['is_licensed'] == True]

# Display the first few rows of the filtered dataframe
licensed_df.head()
```

```python
# Perform value_counts on the 'parent_theme' column of the licensed_df dataframe
parent_theme_counts = licensed_df['parent_theme'].value_counts()

# Display the value counts
parent_theme_counts
```

```python
# Get the number of rows in licensed_df
num_rows_licensed_df = licensed_df.shape[0]

# Display the number of rows
num_rows_licensed_df

the_force = round(parent_theme_counts[0]/num_rows_licensed_df*100)
the_force
```

```python
# Filter merged_df to include only rows where parent_theme is 'Star Wars'
star_wars_df = licensed_df[licensed_df['parent_theme'] == 'Star Wars']

# Display the first few rows of the filtered dataframe
star_wars_df.head()
```
